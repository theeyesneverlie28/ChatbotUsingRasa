```python
import re
import unicodedata
import nltk
from tqdm import tqdm
from string import digits
from bs4 import BeautifulSoup #Nettoyage d'HTML
from spacy.lang.en.stop_words import STOP_WORDS
digits_list = digits


class CleanText:

    def __init__(self, apply_stemming=True):

        #french_stopwords = nltk.corpus.stopwords.words('english')
        #self.stopwords = french_stopwords
        self.stopwords=STOP_WORDS
        self.stemmer = nltk.stem.SnowballStemmer('english')

        self.apply_stemming = apply_stemming


    @staticmethod
    def remove_html_code(txt):
        txt = BeautifulSoup(txt, "html.parser", from_encoding='utf-8').get_text()
        return txt

    @staticmethod
    def convert_text_to_lower_case(txt):
        return txt.lower()

    

    @staticmethod
    def remove_non_letters(txt):
        return re.sub('[^a-z_]', ' ', txt)

    def remove_stopwords(self, txt):
        return [w for w in txt.split() if (w not in self.stopwords)]

    def get_stem(self, tokens):
        return [self.stemmer.stem(token) for token in tokens]

    def apply_all_transformation(self, txt):
        txt = self.convert_text_to_lower_case(txt)
        txt = self.remove_non_letters(txt)
        tokens = self.remove_stopwords(txt)
        if self.apply_stemming:
            tokens = self.get_stem(tokens)
        return tokens
```

```python
mdFile = MdUtils(file_name='data/nlu')
i=-1
for j in range(len(new_table)):
#for j in range(len(1)):

    i+=1
    line=new_table.iloc[j,0]
    text =  "paraphrase: " + line + " </s>"

    max_len = 256

    encoding = tokenizer.encode_plus(text,pad_to_max_length=True, return_tensors="pt")
    input_ids, attention_masks = encoding["input_ids"].to(device), encoding["attention_mask"].to(device)
    

    # set top_k = 50 and set top_p = 0.95 and num_return_sequences = 3
    beam_outputs = model.generate(
        input_ids=input_ids, attention_mask=attention_masks,
        do_sample=True,
        max_length=256,
        top_k=120,
        top_p=0.98,
        early_stopping=True,
        num_return_sequences=5
    )
    final_outputs =[]
    for beam_output in beam_outputs:
        sent = tokenizer.decode(beam_output, skip_special_tokens=True,clean_up_tokenization_spaces=True)
        if sent.lower() != line.lower() and sent not in final_outputs:
            final_outputs.append(sent)
    mdFile.write('## intent: Q'+str(i))
    mdFile.write('\n')
    mdFile.write('- '+line+'\n')

    for final_output in final_outputs:
        mdFile.write('- '+final_output)
        mdFile.write('\n')
    
    
    
    if confi==0:
        L=[]
        L=L+[line]
        ct=CleanText()

        for l in range(1,len(dicti)+1):
            tex=paraphrase_ggtranslate2('en',dicti[l],line)
            s=0
            for k in range(len(L)):
                #if (L[k]).lower()==tex.lower():
                if ct.apply_all_transformation(L[k])==ct.apply_all_transformation(tex):
                    s+=1
            if s==0:
                L=L+[tex.lower()]
                mdFile.write('- '+tex.lower())
                mdFile.write('\n')
        


    
f = open("input/nlu_temp.md", "r")
Lines1 = f.readlines()   
f.close()
for line in Lines1:
    mdFile.write(line)
mdFile.create_md_file()
```
